```
import cv2
import easyocr
import glob
from google.colab.patches import cv2_imshow
```

In the above line of code i have imported the opencv library to work with images, easyocr as our main library that we have clone from the github and the glob module is used to find all the pathnames matching a specified pattern. cv2.imshow is not directly supported in the google colab so i need to import the similar packages from google colab named cv2_imshow.

For further documentation visit the below links:

* Open cv: https://opencv.org/

```
read = easyocr.Reader(['en','ne','hi'])
```

I have tested with only three languages but you can find altogether 59 languages. You can try them out which ever you want according to their language code.
I have used English, Nepali and Hindi for this demo project.

```
path = "/content/drive/My Drive/img/*.*"
for file in glob.glob(path):
   print("\n\n")
   reading_images = cv2.imread(file)
   showing_images = cv2_imshow(reading_images)
   print("\n\n")
   print(showing_images)
   print("\n\n")
   output = read.readtext(file)
   print(str(output))
```

I have uploaded three different photos of English,Nepali and Hindi. In the first line of code i have given the path for my images which are stored in my google drive. Once we mount our google drive(connect our drive with google colab) we can directly access our data that is stored in our google drive. Open cv imread function is used to read the image and convert it into the array of numbers and imshow is used to display the image. The output consists of three differnt parts. our first part with array of number represents the area of boundidng boxes in our text, The second part represents the whole text and third is the confidence associated with the result predicted.

So, we have successfully detected and extracted the text from Images.
